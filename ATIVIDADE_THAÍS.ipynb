```python
#DUPLA
#Lucas Cordeiro Raw
#Lucas Ferreira

import numpy as np
import pandas as pd

from scipy.spatial.distance import squareform, pdist
from sklearn.metrics.pairwise import euclidean_distances
```

```python
pd.options.display.float_format = '{:.2f}'.format

np.set_printoptions(precision=2, suppress=True)

df = pd.read_csv('imrs.csv')
df_salinas = df[df['nome_micro']=='Salinas']
vars = ['mun','perc_lixo','perc_esgoto','esp_vida','tx_mort_inf']

df_salinas = df_salinas[vars].dropna()

df_salinas.head()


```

```python
x = df_salinas.drop(columns=['mun']).to_numpy()
print(x)
```

```python
d = euclidean_distances(x,x)
d
```

```python
df_dist = pd.DataFrame(
    d,
    index=df_salinas['mun'],
    columns=df_salinas['mun']
)
df_dist
```

```python
np.fill_diagonal(d,np.inf)

indice_min = np.unravel_index(np.argmin(d),d.shape)

mun1= df_salinas['mun'].iloc[indice_min[0]]
mun2= df_salinas['mun'].iloc[indice_min[1]]

print("Municípios mais semelhantes:")
print(mun1, "e", mun2)
```

```python
d_max = d.copy()
np.fill_diagonal(d_max, -np.inf)

indice_max = np.unravel_index(np.argmax(d_max), d_max.shape)

mun1_max = df_salinas['mun'].iloc[indice_max[0]]
mun2_max = df_salinas['mun'].iloc[indice_max[1]]

print("\nMunicípios menos semelhantes:")
print(mun1_max, "e", mun2_max)
```
